```python
class AutoEncoder:
    def __init__(self, input_dim, hidden_dims, weight_shared=True):
        self.weight_shared = weight_shared
        self.encoder_params = {}
        self.decoder_params = {}
        self.encoder_hidden_dims = hidden_dims
        self.decoder_hidden_dims = hidden_dims[::-1][1:] + [input_dim]
            
        cur_input_dims = input_dim
        for index, output_dims in enumerate(self.encoder_hidden_dims):
            cur_W_value = np.random.randn(cur_input_dims, output_dims) / np.sqrt(cur_input_dims)
            cur_b_value = np.ones((output_dims))
            self.encoder_params[f"W{index}"] = tf.Variable(cur_W_value, dtype=tf.float32)
            self.encoder_params[f"b{index}"] = tf.Variable(cur_b_value, dtype=tf.float32)
            cur_input_dims = output_dims
        
        for index, output_dims in enumerate(self.decoder_hidden_dims):
            if self.weight_shared:
                self.decoder_params[f"W{index}"] = self.encoder_params[f"W{len(self.encoder_hidden_dims) - index - 1}"]
            else:
                cur_W_value = np.random.randn(cur_input_dims, output_dims) / np.sqrt(cur_input_dims)
                self.decoder_params[f"W{index}"] = tf.Variable(cur_W_value, dtype=tf.float32)
            
            cur_b_value = np.ones((output_dims))
            self.decoder_params[f"b{index}"] = tf.Variable(cur_b_value, dtype=tf.float32)
            cur_input_dims = output_dims

#         l_val = np.sqrt(6) / (np.sqrt(input_dim + hidden_dims[0]))
#         self.W1 = tf.Variable(np.random.uniform(low=-l_val, high=l_val, size=(input_dim, hidden_dims[0])), dtype=tf.float32)
#         self.b1 = tf.Variable(np.ones((hidden_dims[0])), dtype=tf.float32)
#         self.b2 = tf.Variable(np.ones((input_dim)), dtype=tf.float32)
    
    def encoder(self, X):
        hiddens = tf.convert_to_tensor(X, dtype=tf.float32)
        for index in range(len(self.encoder_hidden_dims)):
            hiddens = tf.matmul(hiddens, self.encoder_params[f"W{index}"]) + self.encoder_params[f"b{index}"]
            hiddens = tf.sigmoid(hiddens)
        
#         tensor_X = tf.convert_to_tensor(X, dtype=tf.float32)
#         hiddens2 = tf.matmul(tensor_X, self.W1) + self.b1
        return hiddens
    
    
    def decoder(self, H):
        inputs = tf.convert_to_tensor(H, dtype=tf.float32)
        for index in range(len(self.decoder_hidden_dims)):
            inputs = tf.matmul(inputs, self.decoder_params[f"W{index}"], transpose_b=self.weight_shared) + self.decoder_params[f"b{index}"]
            inputs = tf.sigmoid(inputs)
        
#         tensor_H = tf.convert_to_tensor(H, dtype=tf.float32)
#         inputs2 = tf.matmul(tensor_H, self.W1, transpose_b=True) + self.b2
        return inputs
    
    def predict(self, X):
        hiddens = self.encoder(X)
        inputs = self.decoder(hiddens)
        return inputs
    
    def get_loss(self, X, predictions):
        batch_size = tf.cast(tf.shape(X)[0], dtype=tf.float32)
        sample_loss = tf.reduce_sum((X - predictions) ** 2) / batch_size
        
#         l2_loss = 0
#         for index in range(len(self.encoder_hidden_dims)):
#             l2_loss += tf.reduce_sum(self.encoder_params[f"W{index}"] ** 2)
        
#         for index in range(len(self.decoder_hidden_dims)):
#             l2_loss += tf.reduce_sum(self.decoder_params[f"W{index}"] ** 2)
            
#         loss = 0.5 * (sample_loss + l2_loss)
        
#         pdb.set_trace()
        
        loss2 = sample_loss
        return loss2
    
    def train(self, X, modified_X, keep_rate=1.0, learning_rate=0.05):
        tensor_X = tf.convert_to_tensor(X, tf.float32)
        modified_tensor_X = tf.convert_to_tensor(modified_X, dtype=tf.float32)
        
        with tf.GradientTape(persistent=True) as tape:
            pred = self.predict(modified_tensor_X)
            loss = self.get_loss(tensor_X, pred)
        
        for index in range(len(self.encoder_hidden_dims)):
            grad_Wi = tape.gradient(loss, self.encoder_params[f"W{index}"])
            self.encoder_params[f"W{index}"].assign_sub(learning_rate * grad_Wi)
            
            grad_bi = tape.gradient(loss, self.encoder_params[f"b{index}"])
            self.encoder_params[f"b{index}"].assign_sub(learning_rate * grad_bi)
            
        for index in range(len(self.decoder_hidden_dims)):
            if not self.weight_shared:
                grad_Wi = tape.gradient(loss, self.decoder_params[f"W{index}"])
                self.decoder_params[f"W{index}"].assign_sub(learning_rate * grad_Wi)
                
            grad_bi = tape.gradient(loss, self.decoder_params[f"b{index}"])
            self.decoder_params[f"b{index}"].assign_sub(learning_rate * grad_bi)
        
#         grad_W1 = tape.gradient(loss, self.W1)
#         grad_b1 = tape.gradient(loss, self.b1)
#         grad_b2 = tape.gradient(loss, self.b2)
        
#         self.W1.assign_sub(learning_rate * grad_W1)
#         self.b1.assign_sub(learning_rate * grad_b1)
#         self.b2.assign_sub(learning_rate * grad_b2)
        
#         pdb.set_trace()
```

```python
import matplotlib.gridspec as gridspec


def show_images(images, rows):
    rows = rows
    cols = int(images.shape[0]/rows)
    image_width = image_height = np.ceil(np.sqrt(images.shape[1])).astype(int)
    
    gs = gridspec.GridSpec(rows, cols)
    gs.update(wspace=0.05, hspace=0.05)
    
    for i in range(images.shape[0]):
        axs = plt.subplot(gs[i])
        axs.axis('off')
        axs.set_xticklabels([])
        axs.set_yticklabels([])
        axs.imshow(images[i].reshape([image_width, image_height]))
```

```python
def random_masking(X, keep_rate):
    mask = np.random.binomial(1, keep_rate, X.shape)
    return X * mask


epoch = 100
batch_size = 500
accuracy_size = 500
keep_rate=0.3
max_iter = int(mnist.train.num_examples/batch_size)

autoencoder = AutoEncoder(784, [200])
train_losses = []
val_losses = []

for e in range(epoch):
    for it in range(max_iter):
        batch_x, _ = mnist.train.next_batch(batch_size)
        modified_batch_x = random_masking(batch_x, keep_rate)
        autoencoder.train(X=batch_x, modified_X=modified_batch_x, keep_rate=keep_rate)
    
    train_index = np.random.choice(mnist.train.images.shape[0], accuracy_size)
    train_x = mnist.train.images[train_index]
    modified_train_x = random_masking(train_x, keep_rate)
    pred_train_x = autoencoder.predict(modified_train_x)
    train_loss = autoencoder.get_loss(train_x, pred_train_x)
    
    val_index = np.random.choice(mnist.validation.images.shape[0], accuracy_size)
    val_x = mnist.validation.images[val_index]
    modified_val_x = random_masking(val_x, keep_rate)
    pred_val_x = autoencoder.predict(modified_val_x)
    val_loss = autoencoder.get_loss(val_x, pred_val_x)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    
    
plt.plot(train_losses, label="train losses")
plt.plot(val_losses, label="val losses")
plt.legend()
plt.show()
```

```python
test_index = np.random.choice(mnist.validation.images.shape[0], 5)
test_images = mnist.validation.images[test_index]
modified_test_images = random_masking(test_images, keep_rate)
output_test_images = autoencoder.predict(modified_test_images).numpy()

all_images = np.vstack((test_images, modified_test_images, output_test_images))
show_images(all_images, 3)
```

```python
def random_noising(X, rate):
    noises = np.random.normal(loc=0, scale=rate, size=X.shape)
    return X + noises


epoch = 100
batch_size = 500
accuracy_size = 500
noise_rate=0.7
hidden_layers = [200]
max_iter = int(mnist.train.num_examples/batch_size)

autoencoder2 = AutoEncoder(784, hidden_layers)
train_losses = []
val_losses = []

for e in range(epoch):
    for it in range(max_iter):
        batch_x, _ = mnist.train.next_batch(batch_size)
        modified_batch_x = random_noising(batch_x, noise_rate)
        autoencoder2.train(X=batch_x, modified_X=modified_batch_x, keep_rate=keep_rate)
    
    train_index = np.random.choice(mnist.train.images.shape[0], accuracy_size)
    train_x = mnist.train.images[train_index]
    modified_train_x = random_noising(train_x, noise_rate)
    pred_train_x = autoencoder2.predict(modified_train_x)
    train_loss = autoencoder2.get_loss(train_x, pred_train_x)
    
    val_index = np.random.choice(mnist.validation.images.shape[0], accuracy_size)
    val_x = mnist.validation.images[val_index]
    modified_val_x = random_noising(val_x, noise_rate)
    pred_val_x = autoencoder2.predict(modified_val_x)
    val_loss = autoencoder2.get_loss(val_x, pred_val_x)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    
    
plt.plot(train_losses, label="train losses")
plt.plot(val_losses, label="val losses")
plt.legend()
plt.show()
```

```python
test_index = np.random.choice(mnist.validation.images.shape[0], 5)
test_images = mnist.validation.images[test_index]
modified_test_images = random_noising(test_images, noise_rate)
output_test_images = autoencoder2.predict(modified_test_images).numpy()

all_images = np.vstack((test_images, modified_test_images, output_test_images))
show_images(all_images, 3)
```
